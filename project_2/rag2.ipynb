```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
loader = PyPDFLoader("../data/Evolution_of_AI.pdf")
data = loader.load_and_split()
print(len(data))
data[:5]
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=150,
    length_function=len,
    is_separator_regex=False,
)

text_splitter
```

```python
pages = loader.load_and_split(text_splitter)
```

```python
import os
import google.generativeai as genai
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-pro")

from dotenv import load_dotenv
load_dotenv()
genai.configure(api_key = os.getenv('GEMINI_API'))

embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001")
print(embeddings)
```

```python
import chromadb
chroma_client = chromadb.Client()
collection = chroma_client.create_collection("my_collection2")
```

```python
# Turn chunks into embeddings and add to ChromaDB collection
for i, page in enumerate(pages):
    embedding = embeddings.embed(page.page_content)  # Generate embeddings
    collection.add(
        documents=[page.page_content],
        metadatas=[{"page": i}],
        embeddings=[embedding]
    )
```

```python
import os
import google.generativeai as genai
```

```python
from dotenv import load_dotenv
load_dotenv()
genai.configure(api_key = os.getenv('GEMINI_API'))
```

```python
# This function takes a a sentence as an arugument and return it's embeddings
def get_embeddings(text):
    model = 'models/embedding-001'
    
    # Get the embeddings
    embedding = genai.embed_content(model=model, content=text, task_type="retrieval_document")
    
    print(embedding)
    
    return embedding['embedding']
```

```python
# Get the page_content from the documents and create a new list
content_list = [doc.page_content for doc in docs]
content_list
```

```python
# Send one page_content at a time
embeddings = [get_embeddings(content) for content in content_list]
```

```python
import chromadb
chroma_client = chromadb.Client()
collection = chroma_client.create_collection("my_collection1")
```

```python
for i, doc in enumerate(docs):
    embedding_vector = embeddings.embed_query(doc.page_content)
    collection.add(documents=[{"id": f"doc_{i}", "text": doc.page_content}], embeddings=[embedding_vector])
```

```python
# Add documents to ChromaDB collection
for i, doc in enumerate(docs):
    embedding_vector = embeddings.embed_query(doc.page_content)
    collection.add(documents=[{"id": f"doc_{i}", "text": doc.page_content}], embeddings=[embedding_vector])
```
